```python
gamma = np.linspace(0, np.pi, 19)
gamma_grad = np.linspace(0, 180, 19)
w1 = np.cos(gamma)
w2 = np.sin(gamma)
plt.scatter(w1,w2)
plt.xlabel('w1')
plt.ylabel('w2')
```

```python
w= np.array( list(zip(w1,w2)))
x = np.array(list(zip(data['x.1'], data['x.2'])))
y = np.array(data['y'])


def transfer_function (w,x):
    y_n =[]
    if np.dot(w.T,x) >= 0:
        y_n.append(1)
    else :
        y_n.append(0)
    return y_n


def test (w,x):
    y_final = []
    for i in  w :
        y_array = []
        for j in x:
            y_array.append(transfer_function(i,j))
        y_final.append(y_array)
    result = []
    for k in  y_final :
        result.append(200 - np.count_nonzero (k -y[k]))
    return np.array(result)/2
percentage = test (w,x)
    
    
```
